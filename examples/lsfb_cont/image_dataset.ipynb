```python
from torchvision import transforms
from torch.utils.data import DataLoader
from lsfb_dataset.datasets.lsfb_cont.image_folder import ImageDataset
```

```python
transform = transforms.ToTensor()
dataset = ImageDataset('T:/datasets/lsfb_cont/images', features=['right_hand', 'left_hand'], transform=transform, categories_nb=4, categories_offset=2)
len(dataset)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

dataloader = DataLoader(dataset, batch_size=32, shuffle=True, num_workers=0)

for item in dataloader:
    print(item.keys())

    print(dataset.class_names[item['right_hand_label'][0]])
    plt.imshow( item['right_hand'][0].permute(1, 2, 0) )
    plt.show()

    print(dataset.class_names[item['left_hand_label'][0]])
    plt.imshow( item['left_hand'][0].permute(1, 2, 0) )
    plt.show()
    break
```
